#### Write a program in C# Sharp to separate odd and even integers in separate arrays. 

```python
array_count = int(input("The number elements to be stored in the array: "))
array_even=[]
array_odd=[]
array_total=[]
i=0
while (i<array_count):
    input_element= input("element-" + str(i)+ ":")
    if input_element.isnumeric():
        array_total.append(int(input_element))
        i=i+1
    else:
        print("Please enter a valid number")

for i in array_total:
    if i%2==0:
        array_even.append(i)

    else:
        array_odd.append(i)
print("The even elements are: ")
for i in array_even:
    print(i,end=" ")
print("\n")
print("The odd elements are: ")
for i in array_odd:
    print(i,end=" ")
```

```python
def inside(x, y, x1, y1, x2, y2):
    
    if x > x1 and x < x2 and y > y1 and y < y2:
        return True
    else:
        return False
inside(-1,-1,0,0,2,3)
```

#### 16. You can turn a word into pig-Latin using the following two rules (simplified):
#### • If the word starts with a consonant, move that letter to the end and append 'ay'. For example, 'happy' becomes 'appyhay' and 'pencil' becomes 'encilpay'.
#### • If the word starts with a vowel, simply append 'way' to the end of the word. For example, 'enter' becomes 'enterway' and 'other' becomes 'otherway' . For our purposes, there are 5 vowels: a, e, i, o, u (so we count y as a consonant).
#### Write a function pig() that takes a word (i.e., a string) as input and returns its pigLatin form. Your function should still work if the input word contains upper case characters. Your output should always be lower case however

```python
def pig(word):
    if word.isalpha():
        if word[0].lower() in 'aeiou':
            word= word.lower()+"way"
            return word.lower()

        else:
            new_word= word[1:].lower()+word[0].lower()+"ay"
            return new_word
    else:
        print("Plesae enter a valid string")
user_input= input("Enter the word to check: ")
print(pig(user_input))
```
